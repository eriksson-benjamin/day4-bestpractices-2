    I will plot a 2D histogram of some random data in this notebook

```python
import numpy as np
import matplotlib
matplotlib.use('Agg')
%matplotlib inline
import matplotlib.pyplot as plt
```

Get some random data using numpy.random.normal()

```python
n_samples = 100000
y_rand = np.random.normal(size = n_samples)
x_rand = np.random.normal(size = n_samples)
```

Plot the randomly generated data in a 2D histogram.

```python
fig = plt.figure()
ax = fig.add_subplot(111)

# Define bins and plot
x_bins = np.arange(-4, 4, 0.1)
y_bins = np.arange(-4, 4, 0.1)
hist = ax.hist2d(x_rand, y_rand, bins = [x_bins, y_bins])

# Set axis labels and colour bar
ax.set_xlabel('x')
ax.set_ylabel('y')
fig.colorbar(hist[3], ax = ax)
ax.set_title('Normally distributed 2D data')
```
